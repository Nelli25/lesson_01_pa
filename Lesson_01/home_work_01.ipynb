#### 1.2. На складе лежат разные фрукты в разном количестве. Написать функцию, которая на вход принимает любое количество названий фруктов и их количество, а возвращает общее количество фруктов на складе
Примеры/Тесты:

        <finction_name>(apple=2, banana=12, pear=9, peach=5) -> 28

````python

def total_fruits(*args):
    """
    Функция принимает любое количество названий фруктов и их количество,
    а возвращает общее количество фруктов на складе.
    """
    total = 0
    for fruit in args:
        total += fruit
    return total
apples = 2
bananas = 12
pears = 9
peaches = 5

total = total_fruits(apples, bananas, pears, peaches)
print(f"Общее количество фруктов на складе: {total}")
````

#### 1.3. Дан список с затратами на рекламу. Но в данных есть ошибки, некоторые затраты имеют отрицательную величину. Удалите такие значения из списка и посчитайте суммарные затраты. Используйте list comprehensions.
Примеры/Тесты:

        [-12, 100, 0, 125, -90, 0, 345, 655, -1, 0, 200] -> [100, 0, 125, 0, 345, 655, 0, 200]

```(*)``` **Усложнение.** Удалите отрицательные значения из списка, а нулевые значения замените на None
Примеры/Тесты:

        [-12, 100, 0, 125, -90, 0, 345, 655, -1, 0, 200] -> [100, None, 125, None, 345, 655, None, 200]

````python
ad_costs = [-12, 100, 0, 125, -90, 0, 345, 655, -1, 0, 200]

# Используем list comprehensions для удаления отрицательных значений
ad_costs = [cost for cost in ad_costs if cost >= 0]

# Считаем сумму затрат
total_costs = sum(ad_costs)

print(f"Суммарные затраты на рекламу: {total_costs}")
````

````python
# Усложненный вариант

ad_costs = [-12, 100, 0, 125, -90, 0, 345, 655, -1, 0, 200]

# Используем list comprehensions для удаления отрицательных значений и замены нулевых на None
ad_costs = [cost if cost > 0 else None for cost in ad_costs]

print(ad_costs)
````

        Ответ: 25098
2. Найдите выручку компании в зависимости от месяца. Напишите функцию, которая на вход принимает списки с датами и с выручкой, а на выходе вернет словарь, где ключи - это месяцы, а значения - это выручка.
Примеры/Тесты:

        <finction_name>(dates, prices) -> {'09': 25647, '12': 45452, '10': 28645, '11': 25098}

*Примечание:* 
- Для извлечения месяца вы можете воспользоваться пакетом datetime или извлекать его как строку
- Для работы "синхронно" с двумя последовательностями вы можете использовать индекс или функцию zip. Попробуйте оба варианта.
- Используйте аннотирование типов при определении функции

````python
from typing import List, Dict
import datetime

dates = ['2021-09-14', '2021-12-15', '2021-09-08', '2021-12-05', '2021-10-09', '2021-09-30', '2021-12-22', '2021-11-29', '2021-12-24', '2021-11-26', '2021-10-27', '2021-12-18', '2021-11-09', '2021-11-23', '2021-09-27', '2021-10-02', '2021-12-27', '2021-09-20', '2021-12-13', '2021-11-01', '2021-11-09', '2021-12-06', '2021-12-08', '2021-10-09', '2021-10-31', '2021-09-30', '2021-11-09', '2021-12-13', '2021-10-26', '2021-12-09']
amounts = [1270, 8413, 9028, 3703, 5739, 4095, 295, 4944, 5723, 3701, 4471, 651, 7037, 4274, 6275, 4988, 6930, 2971, 6592, 2004, 2822, 519, 3406, 2732, 5015, 2008, 316, 6333, 5700, 2887]

# Найдем выручку компании в ноябре
november_revenue = sum([amounts[i] for i in range(len(dates)) if datetime.datetime.strptime(dates[i], '%Y-%m-%d').month == 11])
print("Выручка компании в ноябре:", november_revenue)

# Напишем функцию для нахождения выручки компании по месяцам
def revenue_by_month(dates: List[str], amounts: List[int]) -> Dict[str, int]:
    revenue_dict = {}
    for i in range(len(dates)):
        month = datetime.datetime.strptime(dates[i], '%Y-%m-%d').month
        if month in revenue_dict:
            revenue_dict[month] += amounts[i]
        else:
            revenue_dict[month] = amounts[i]
    return revenue_dict

# Используем функцию для нахождения выручки компании по месяцам
monthly_revenue = revenue_by_month(dates, amounts)
print("Выручка компании по месяцам:", monthly_revenue)


# Выручка компании в ноябре: 22035
# Выручка компании по месяцам: {9: 36379, 12: 29207, 10: 17932, 11: 22035}
````
