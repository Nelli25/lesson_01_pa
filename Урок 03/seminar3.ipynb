- Считать данные с помощью pandas
- Вывести на экран первые 5 строк
- Создать новый признак Cpu_Company, который будет содержать только название фирмы, которая произвела CPU (Intel, AMD и т.д)
- Создать новый признак Memory_Amount, который будет содержать только количество Gb памяти без указания типа носителя (128, 1024, 256 и т.д.)
- Создать новый признак Memory_Type, который будет содержать только тип носителя (HDD, SDD и т.д.)
- Удалите признаки Memory и ScreenResolution

```python
import pandas as pd
```

```python
df = pd.read_csv('laptop_price.csv', encoding='1251')
```

```python
df.head()
```

```python
def vstavka(stroka):
    return stroka.split()[0]
```

```python
df['Cpu_Company'] = df['Cpu'].apply(vstavka)
```

```python
df.head()
```

```python
df['Cpu_Company'].value_counts()
```

```python
df[['Cpu_Company', 'Cpu']]
```

```python
#Создать новый признак Memory_Amount, 
#который будет содержать только количество Gb памяти без указания типа носителя (128, 1024, 256 и т.д.)
```

```python
def sumMem(stroka):
    return int(float(stroka.split()[0][:-2]))
```

```python
df['Memory_Amount'] = df['Memory'].apply(sumMem)
```

```python
df.head()
```

```python
df['Memory'].value_counts()
```

```python
sumMem('256GB SSD +  1.0TB Hybrid')
```

```python
sumMem('2TB HDD')
```

```python
#Создать новый признак Memory_Type, который будет содержать только тип носителя (HDD, SDD и т.д.)
```

```python
def typeMem(stroka1):
    return stroka1.split()[1]
```

```python
df['Memory_Type'] = df['Memory'].apply(typeMem)
```

```python
typeMem('256GB SSD +  1.0TB Hybrid')
```

```python
df.head()
```

```python
df['Memory_Type'].value_counts()
```

```python
#Удалите признаки Memory и ScreenResolution
```

```python
df.drop(columns=['Memory', 'ScreenResolution'], inplace=True)
```

```python
df.head()
```
